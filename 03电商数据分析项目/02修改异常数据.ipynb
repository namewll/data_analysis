```python
import pandas as pd
shop=pd.read_excel("shop_new.xlsx",None)
order=shop['订单信息']
customer=shop['顾客信息']
good=shop['货物信息']
customer
```

```python
for id in customer.客户ID:
    # c1=customer.loc[customer.客户ID==id,"订单数"].values[0]
    c1=customer[customer.客户ID==id]['订单数'].values[0]
    c2=order[(order.客户ID==id)&(order.订单状态==1)].shape[0]
    if c1!=c2:
        customer.loc[customer.客户ID==id,"订单数"]=c2
customer
```

```python
writer=pd.ExcelWriter("shop_new2.xlsx")
customer.to_excel(writer,sheet_name="顾客信息",index=False)
order.to_excel(writer,sheet_name="订单信息",index=False)
good.to_excel(writer,sheet_name="货物信息",index=False)
writer._save()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import font_manager
font=font_manager.FontProperties(fname='STCAIYUN.TTF')
```

```python
plt.rcParams['font.family']=['Microsoft YaHei']
plt.rcParams['axes.unicode_minus']=False
```

```python
a1=pd.read_excel("餐饮经营分析：起死回生的奶茶店.xlsx")
a1
```

```python
a2=a1.set_index('年份')
a2
```

```python
a3=a2.T
a3
```

```python
x=a3.index.values
x
```

```python
y1=a3['产品销售收入（万）']
y2=a3['合作店收入（万）']
y3=a3['总收入（万）']
y3
```

```python
plt.figure(figsize=(5,4))
plt.plot(x,y1,label='产品销售收入',color='r')
plt.plot(x,y2,label='合作店收入',color='b')
plt.plot(x,y3,label='总收入',color='g')
plt.title('奶茶企业收入和各类收入稳定趋势',font=font)
plt.legend()
plt.show()
```

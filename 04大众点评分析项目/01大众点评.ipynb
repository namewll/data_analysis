```python
import pandas as pd
# 一次性导入所有图例
from pyecharts.charts import * 
# 导入环境配置代码 可视化的依赖
from pyecharts.globals import CurrentConfig, NotebookType  
CurrentConfig.NOTEBOOK_TYPE = NotebookType.JUPYTER_LAB
# 配置选项 
from pyecharts import options as opts
# 主题的类型
from pyecharts.globals import ThemeType
```

```python
a1=pd.read_excel('./大众点评.xlsx')
a1
```

```python
a1.info()
```

```python
a1.describe()
```

```python
a1.dropna(axis=0,inplace=True)
a1.info()
```

```python
import re
```

```python
a1['口味评分']=a1['口味评分'].apply(lambda x:re.findall('\d+\.?\d*',x)[0]).astype('float')
a1
```

```python
a1.info()
```

```python
a1['人均消费']=a1['人均消费'].apply(lambda x:re.findall('\d+',x)[0]).astype('int')
a1.info()
```

```python
a2=a1.groupby('城市')['人均消费'].mean().sort_values(ascending=False).head(10)
a2
```

```python
bar = (
    # 主题
    Bar(init_opts=opts.InitOpts(theme='white',bg_color='rgba(255,182,193,0.4)'))
    .add_xaxis(a2.index.tolist())
    .add_yaxis("人均消费", a2.values.tolist())
    .set_global_opts(title_opts=opts.TitleOpts(title="个城市人均消费", subtitle="副标题"))
)
bar.load_javascript()
bar.render('./图例/1.html')
```

```python
a3=a1.groupby('城市')['门店整体评分'].sum().sort_values(ascending=False).head(5)
a4=a3.rename(index={'上海':'上海市','福州':'福建省','深圳':"广东省",'沈阳':'辽宁省','苏州':'江苏省'})
```

```python
map1 = Map()
map1.add("",[z for z in zip(a4.index.tolist(),[15,114,236,610,895])],maptype='china')
map1.set_global_opts(
    title_opts = opts.TitleOpts(title='大众点评国内各地美食评分分布图'),
    visualmap_opts = opts.VisualMapOpts(max_=900)
)
map1.load_javascript()
map1.render('./图例/3.html')
```

```python
def parse(x):
    if x<50:
        return "50元以下"
    elif x<150:
        return "50元-150元之间"
    elif x<300:
        return "150元-300元之间"
    elif x<400:
        return "300元-400元之间"
    else:
        return '400元以上' 
a1['人均消费'].apply(lambda x:parse(x))
```

```python
pd.cut(a1['人均消费'],bins=[0,50,150,300,400,500])
```

```python
a1['人均消费区间']=a1['人均消费'].apply(lambda x:parse(x))
a1
```

```python
s=a1['人均消费区间'].value_counts()
s
```

```python
bar3 = Bar() 
bar3.add_xaxis(s.index.tolist())
bar3.add_yaxis('', s.values.tolist()) 
bar3.set_global_opts(title_opts=opts.TitleOpts(title='大众点评人均消费不同价格区间情况'),
                     visualmap_opts=opts.VisualMapOpts(max_=65)) 
bar3.load_javascript()
bar3.render('./图例/4.html')
```

```python
data_pair=[data_pair for data_pair in zip(s.index.tolist(),s.values.tolist())]
pie1 = Pie()
pie1.add('', data_pair, radius=['35%', '60%'])
pie1.set_global_opts(title_opts=opts.TitleOpts(title='大众点评不同价格区间人均消费的表现'), 
                     legend_opts=opts.LegendOpts(orient='vertical', pos_top='15%', pos_left='2%'))
pie1.set_series_opts(label_opts=opts.LabelOpts(formatter="{b}:{d}%"))
pie1.set_colors(['#EF9050', '#3B7BA9', '#6FB27C', '#FFAF34', '#D8BFD8', '#00BFFF', '#7FFFAA'])
pie1.load_javascript()
pie1.render('./图例/5.html')
```
